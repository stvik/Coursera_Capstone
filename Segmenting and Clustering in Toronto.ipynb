```python
import requests
from bs4 import BeautifulSoup
```

Scraping the data from Wikipedia using BeautifulSoup:

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

Create a table called My_table from the table of Neighborhoods on Wikipedia by finding the class 'wikitable sortable'.

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
A=[]
B=[]
C=[]


for row in My_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

Create a dataframe using the information stored in A, B, and C.

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postal Code'])
df['Bourough']=B
df['Neighborhood']=C
df.head()
```

Remove all 'not assigned' 

```python
df = df[df.Bourough != 'Not assigned']
```

```python
df.replace('Not assigned\n','Queen\'s Park')
df.head()
```

Combine neighborhoods with the same postalcode into one row

```python
df1 = df.groupby(['Postal Code','Bourough']).agg({'Neighborhood':lambda x: list(x)}).reset_index()
df1.head()
```

Find the shape of the dataframe

```python
df1.shape
```

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python

map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, borough, neighborhood in zip(finaldf['Latitude'], finaldf['Longitude'], finaldf['Bourough'], finaldf['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
